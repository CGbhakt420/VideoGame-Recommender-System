```python
content_df = data[['id', 'name', 'platforms', 'developers', 'genres', 'publishers']]
```

```python
def make_list(obj):
    obj = obj
    tags = []
    devs = obj['developers'].split('||')
    genres = obj['genres'].split('||')
    pubs = obj['publishers'].split('||')

    for tag in genres:
        tags.append(tag)
    for tag in devs:
        tags.append(tag)
    for tag in pubs:
        tags.append(tag)
    return tags
```

```python
tags_dict = {}
tags_dict['tags'] = []
for index, row in content_df.iterrows():
    tag = make_list(row)
    tags_dict['tags'].append(tag)

tags_dict
```

```python
for i in range(len(tags_dict['tags'])):
    tags_dict['tags'][i] = list(set(tags_dict['tags'][i]))

tags_dict
```

```python
for i in range(len(tags_dict['tags'])):
    for j in range(len(tags_dict['tags'][i])):
        tags_dict['tags'][i][j] = tags_dict['tags'][i][j].replace(" ", "")

tags_dict
```

```python
tags = pd.Series(tags_dict['tags'], name='Tags')
tags = tags.to_frame()

tags
```

```python
content_df = content_df.reset_index()
content_df = pd.concat([content_df, tags], axis=1)
content_df['Tags'] = content_df['Tags'].apply(lambda x: " ".join(x))
content_df['Tags'] = content_df['Tags'].apply(lambda x: x.lower())
```

```python
content_df.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500, stop_words='english')

vector = cv.fit_transform(content_df['Tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similar = cosine_similarity(vector)
similar
```

```python
def Recommendation(name):
    idx = content_df[content_df['name'] == name].index[0]
    distances = sorted(list(enumerate(similar[idx])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(content_df.iloc[i[0]]['name'])
```

```python
Recommendation("HITMAN")
```
